```python
# 1. Practice Problem 1 — Tensor Operations

import torch

# 1. Create a tensor of size 3 × 4 with random values
tensor_a = torch.rand(3,4)

# 2. Compute the element-wise square of the tensor.
tensor_squared = torch.pow(tensor_a, 2)

# 3. Generate a vector of size 4 and compute the matrix-vector product.
tensor_b = torch.rand(4)
mul_tensor = torch.matmul(tensor_a, tensor_b)

# 4. Print the tensor size and number of elements
print(f"size: {tensor_a.size()}")
print(f"number of element: {tensor_a.numel()}")
```

```python
# 2. Practice Problem 2 — Gradient Computation

import torch

# 1. Define a tensor xxx with requires_grad=True.
x = torch.tensor(10.0, requires_grad=True)

# 2. Define the function f(x) = x2 + 3x
f = x**2+3*x

# 3. Compute the gradient using backward().
f.backward()

# 4. Print the gradient value.
print(f"gradient value: {x.grad}")

```
